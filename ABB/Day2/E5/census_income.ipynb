```python
temp1 = df['workclass'].value_counts(ascending=True)
print ('Frequency Table for workclass:') 
print (temp1)

 
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(121)
ax1.set_xlabel('workclass')
ax1.set_ylabel('Count of people')
ax1.set_title("people by workclass")
temp1.plot(kind='bar')
```
